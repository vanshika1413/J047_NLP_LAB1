```python
base_names_path = '/kaggle/input/datadata/base_names.csv' 
name_variations_path = '/kaggle/input/datadata/name_variations.csv' 
base_names = pd.read_csv(base_names_path)
name_variations = pd.read_csv(name_variations_path)

# Extract lists of names from both dataframes
base_names_list = base_names['Base_Name'].tolist()
variations_list = name_variations['Variation'].tolist()
```

```python
# Define a function to perform fuzzy matching using difflib
def fuzzy_match_names_difflib(variations, base_names_list, threshold=0.8):
    matched_results = []
    for name in variations:
        # Use difflib.get_close_matches to find the best match in base_names_list for each variation
        best_matches = difflib.get_close_matches(name, base_names_list, n=1, cutoff=threshold)
        if best_matches:
            best_match = best_matches[0]
            # Calculate similarity ratio
            similarity_score = difflib.SequenceMatcher(None, name, best_match).ratio() * 100
            matched_results.append((name, best_match, similarity_score))
        else:
            matched_results.append((name, None, 0))  # No match found
    return matched_results

# Perform fuzzy matching
matched_names = fuzzy_match_names_difflib(variations_list, base_names_list)


matched_df = pd.DataFrame(matched_names, columns=['Variation', 'Best_Match', 'Similarity_Score'])
matched_df.to_csv('fuzzy_matched_names.csv', index=False)
print(matched_df.head())
```
